### DeepFM: Regression

#### Regression approach in predicting FCR_7days using DEEPFM

```python
sparse_features = ["IxnID","Agent_Greeting","Customer_ANI","CallReason","result_reason","IVR_NAME","Priority","customer_acw_duration","Skill","business_result","virtual_queue","AHT","technical_result","days","is_weekend"]
target = ['FCR_7days']

# 1.Label Encoding for sparse features
for feat in sparse_features:
    lbe = LabelEncoder()
    copy_dataset[feat] = lbe.fit_transform(copy_dataset[feat])
# 2.count #unique features for each sparse field
fixlen_feature_columns = [SparseFeat(feat, copy_dataset[feat].nunique())
                              for feat in sparse_features]
linear_feature_columns = fixlen_feature_columns
dnn_feature_columns = fixlen_feature_columns
fixlen_feature_names = get_fixlen_feature_names(linear_feature_columns + dnn_feature_columns)

# 3.generate input data for model
train, test = train_test_split(copy_dataset, test_size=0.2)
train_model_input = [train[name].values for name in fixlen_feature_names]
test_model_input = [test[name].values for name in fixlen_feature_names]

# 4.Define Model,train,predict and evaluate
model = DeepFM(linear_feature_columns, dnn_feature_columns,use_fm=True, dnn_hidden_units=(128, 128), l2_reg_linear=1e-05, l2_reg_dnn=0, init_std=0.0001, seed=1024, dnn_dropout=0, dnn_activation='relu', dnn_use_bn=False, task='regression')
model.compile("adam", "mse", metrics=['mse'], )
history = model.fit(train_model_input, train[target].values,
                        batch_size=256, epochs=10, verbose=2, validation_split=0.2)
```

```python
pred_ans_reg = model.predict(test_model_input, batch_size=256)
result_dataFrame=test[["IxnID","Agent_Greeting","Customer_ANI","CallReason","result_reason","IVR_NAME","customer_acw_duration","Priority","Skill","business_result","virtual_queue","AHT","technical_result","days","is_weekend","FCR_7days"]]
result_dataFrame["Predicted_FCR_7days"]=pd.DataFrame(pred_ans_reg).values.round()
result_dataFrame.head(10)                        
```

## Balancing(Without Replacement)

```python
def train_test_balanced1(interested_features,test_size):
    """Method: Here in order to balance the 0s and 1s in the test and train we sample without replacement.
       Prodcedure:
         1. Seperate the interested dataframe into two dataframes one containing those rows having FCR_7days equals 1, 
             say One_FCR and the other one having rows where FCR_7days is 0, say zero_FCR.
             
         2. Draw sample of size len(zero_FCR) without replacement from One_FCR and then club this sampled dataframe and
            our zero_FCR into one dataframe which will be used for the train and test split.
         
         3. Draw equal samples from One_FCR and zero_FCR without replacement with sample size rounded around half of the test size
            in order to ensure that the proportion of FCR_7days = 1 and 0 are same and club these two set of rows
            together, this will be the test set.
         
         4. Remove those rows from One_FCR and zero_FCR which are present in the test set.
             
         5. Club the remaining rows in One_FCR and zero_FCR, this will be the train set.
            
    """
    
    # Seperating the rows with FCR_7days=1 in one dataframe and that with zero in another
    One_FCR=interested_features.loc[interested_features['FCR_7days'] == 1]
    zero_FCR=interested_features.loc[interested_features['FCR_7days'] == 0]
    
    # Reducing the number of ones by drawing a sample of size equal to the number of rows having FCR_7days=0 from One_FCR
    New_One_FCR=One_FCR.sample(len(zero_FCR),replace=False)# New One_FCR
    
    # Below is the dataframe used for doing the train and test split by clubbing together One_FCR and zero_FCR
    train_test=pd.concat([New_One_FCR,zero_FCR],axis=0)# Proportion FCR_7days=1 and 0 are the same here
    
    # Computing the size of the part having FCR_7days=0 and that having 1 in the test set
    zero_test_size=round(round(test_size*len(train_test))/2)
    one_test_size=zero_test_size
    
    # Test part containing FCR_7days=0
    zero_FCR_sample_test=zero_FCR.sample(zero_test_size,replace=False)
    # Test_part containing FCR_7days=1
    One_FCR_sample_test=New_One_FCR.sample(one_test_size,replace=False)
    
    # Forming the test dataframe by concatinating zero_FCR_sample_test and One_FCR_sample_test
    test=pd.concat([zero_FCR_sample_test,One_FCR_sample_test],axis=0)
    
    # Train part containing FCR_7days=1
    one_sample_train=pd.concat([New_One_FCR,One_FCR_sample_test]).drop_duplicates(keep=False)
    # Train part containing FCR_7days=0
    zero_sample_train=pd.concat([zero_FCR,zero_FCR_sample_test]).drop_duplicates(keep=False)
    
    # Forming the train dataframe by concatinating zero_sample_train and one_sample_train
    train=pd.concat([zero_sample_train,one_sample_train],axis=0)
    return train,test
```

## Balancing(With Replacement)

```python
 def train_test_balanced2(interested_features,test_size):
   
    """Method: Here in order to balance the 0s and 1s in the test and train we sample with replacement.
       Prodcedure:
         1. Seperate the interested dataframe into two dataframes one containing those rows having FCR_7days equals 1, 
             say One_FCR and the other one having rows where FCR_7days is 0, say zero_FCR.
             
         2. Sample the test part containing FCR_7days=0 from zero_FCR with replcament and test part containing FCR_7days=1 without replacement
         
         4. Remove those rows from One_FCR and zero_FCR which are present in the test set.
             
         5. Club the remaining rows in One_FCR and zero_FCR, this will be the train set.
            
    """
    
    # Seperating the rows with FCR_7days=1 in one dataframe and that with zero in another
    One_FCR=interested_features.loc[interested_features['FCR_7days'] == 1]
    zero_FCR=interested_features.loc[interested_features['FCR_7days'] == 0]
    
    # Computing the size of the part having FCR_7days=0 and that having 1 in the test set
    zero_test_size=round(test_size*len(One_FCR))# Since actual size of the dataframe to be used is twice the number of rows in One_FCR as we are trying to keep equal number of rows having FCR_7days=0 and 1.
    one_test_size=zero_test_size
    
    # Test part containing FCR_7days=0
    zero_FCR_sample_test=zero_FCR.sample(zero_test_size,replace=True)
    # Test_part containing FCR_7days=1
    One_FCR_sample_test=One_FCR.sample(one_test_size,replace=False)
    
    # Forming the test dataframe by concatinating zero_FCR_sample_test and One_FCR_sample_test
    test=pd.concat([zero_FCR_sample_test,One_FCR_sample_test],axis=0)
    
    # Train part containing FCR_7days=1
    one_sample_train=pd.concat([One_FCR,One_FCR_sample_test]).drop_duplicates(keep=False)
    # Train part containing FCR_7days=0
    zero_sample_remain=pd.concat([zero_FCR,zero_FCR_sample_test]).drop_duplicates(keep=False)
    zero_sample_re=zero_sample_remain.sample(len(one_sample_train)-len(zero_sample_remain),replace=True)
    zero_sample_train=pd.concat([zero_sample_remain,zero_sample_re],axis=0)
    
    # Forming the train dataframe by concatinating zero_sample_train and one_sample_train
    train=pd.concat([zero_sample_train,one_sample_train],axis=0)
    return train,test
```
